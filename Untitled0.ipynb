```python
pop_seoul = pd.read_excel('pop_in_seoul.xlsx', usecols=[1,3,6,9,13], header=2)
```

```python
pop_seoul.head()
```

```python
pop_seoul.columns = ['구별','인구수','한국인','외국인','고령자']
```

```python
pop_seoul.head()
```

```python
pop_seoul.set_index('구별', inplace=True)
```

```python
pop_seoul.head()
```

```python
pop_seoul.drop('합계', inplace=True)
```

```python
pop_seoul.head()
```

```python
pop_seoul['한국인비율(%)']=(pop_seoul.한국인/pop_seoul.인구수)*100
```

```python
pop_seoul.head()
```

```python
pop_seoul['외국인비율(%)']=(pop_seoul.외국인/pop_seoul.인구수)*100
```

```python
pop_seoul.head()
```

```python
pop_seoul['고령자비율(%)']=(pop_seoul.고령자/pop_seoul.인구수)*100
```

```python
pop_seoul.head()
```

```python
pop_seoul.sort_values(by='인구수', ascending=False)
```

```python
pop_seoul.sort_values(by='인구수', ascending=True).head()
```

```python
pop_seoul.sort_values(by='외국인', ascending=False)
```

```python
pop_seoul.sort_values(by='고령자비율(%)', ascending=True)
```
